Bu notebook her öğrenme kazanımının hangi Bloom düzeyine atandığını ve neden atandığını tablo halinde gösterir.

### Bloom Taksonomisi — 6 Düzey ve Katsayılar

| Düzey | Ad | Katsayı | Örnek Fiiller |
|-------|----|---------|---------------|
| L1 | Hatırlama | 1.00 | tanımlar, listeler, adlandırır |
| L2 | Anlama | 1.10 | açıklar, özetler, sınıflandırır |
| L3 | Uygulama | 1.20 | uygular, hesaplar, çözer |
| L4 | Analiz | 1.30 | analiz eder, inceler, sorgular |
| L5 | Değerlendirme | 1.40 | değerlendirir, eleştirir, karar verir |
| L6 | Yaratma | 1.50 | tasarlar, geliştirir, oluşturur |

```python
import pandas as pd

df = pd.read_excel("../data/processed/master_bloom.xlsx", dtype={"Katalogid": str})
print(f"Yüklendi: {len(df)} satır, {len(df.columns)} sütun")
```

### Ders Bazlı Bloom Özet Tablosu

Her ders için dominant Bloom düzeyi, ortalama düzey, K_bilissel katsayısı ve gerekçesi gösterilir.

```python
ozet = df[[
    "Katalogid", "DersAdi", "Fakulte", "alan_grubu",
    "AKTSKredi", "n_ogrenme_kazanim",
    "bloom_dominant_level", "bloom_avg_level",
    "bloom_max_level", "K_bilissel", "bloom_justification"
]].copy()

ozet["Bloom_Etiketi"] = ozet["bloom_dominant_level"].map({
    1: "L1 — Hatırlama",
    2: "L2 — Anlama",
    3: "L3 — Uygulama",
    4: "L4 — Analiz",
    5: "L5 — Değerlendirme",
    6: "L6 — Yaratma"
})

ozet = ozet.rename(columns={
    "DersAdi": "Ders Adı",
    "Fakulte": "Fakülte",
    "alan_grubu": "Alan Grubu",
    "AKTSKredi": "AKTS",
    "n_ogrenme_kazanim": "Kazanım Sayısı",
    "bloom_dominant_level": "Dominant Düzey",
    "bloom_avg_level": "Ort. Düzey",
    "bloom_max_level": "Max Düzey",
    "K_bilissel": "K_bilissel",
    "bloom_justification": "Gerekçe",
    "Bloom_Etiketi": "Bloom Etiketi"
})

display(ozet.head(20))
print(f"Toplam ders: {len(ozet)}")
```

### Bloom Düzeyi Dağılımı

```python
dagilim = df["bloom_dominant_level"].map({
    1: "L1 — Hatırlama",
    2: "L2 — Anlama",
    3: "L3 — Uygulama",
    4: "L4 — Analiz",
    5: "L5 — Değerlendirme",
    6: "L6 — Yaratma"
}).value_counts().reset_index()
dagilim.columns = ["Bloom Düzeyi", "Ders Sayısı"]
dagilim["Oran (%)"] = (dagilim["Ders Sayısı"] / len(df) * 100).round(1)
display(dagilim)
```

### Fakülte Bazlı Ortalama Bloom Düzeyi

```python
fakulte_bloom = df.groupby("Fakulte").agg(
    Ders_Sayisi=("Katalogid", "count"),
    Ort_Bloom=("bloom_avg_level", "mean"),
    Ort_K_bilissel=("K_bilissel", "mean")
).round(3).reset_index()
fakulte_bloom.columns = ["Fakülte", "Ders Sayısı", "Ort. Bloom Düzeyi", "Ort. K_bilissel"]
display(fakulte_bloom)
```

### Önemli Bulgu

%68 oranında ders **L1 (Hatırlama)** düzeyinde kalmaktadır.  
Bu, EBS sistemindeki öğrenme kazanımlarının büyük çoğunluğunun Bloom taksonomisine uygun yazılmadığını göstermektedir.  
Bu bulgu projenin temel tezini **(H4 hipotezi)** desteklemektedir.

```python
ozet.to_excel("../data/processed/bloom_duzey_tablo.xlsx", index=False)
print("bloom_duzey_tablo.xlsx kaydedildi.")
print(f"Satır: {len(ozet)}, Sütun: {len(ozet.columns)}")
```
